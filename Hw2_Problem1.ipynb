## Problem 1:
Find three simple bivariate functions f1, f2, and f3 and three points p1, p2, p3 such that 

*   p1 is a minimum of f1
*   p2 is a maximum of f2
*   p3 is neither a minimum nor maximum of f3, but the gradient of f3 at p3 is the zero vector.

Use (a) three-dimensional plots as in three_dimensional_plotting.ipynb and (b) density plots and (c) contour plots as in density_and_contour_plots.ipynb. Indicate clearly the points p1, p2, and p3 in all plots.

###**Function f1 where p1 is a minimum of f1:**

```
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```
def f1(x,y):
  return x **  2 + y ** 2

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f1(X, Y)
```

```
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(X, Y, Z, 50, cmap = 'winter')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
z_min = np.min(Z)
x_min,y_min = np.unravel_index(np.argmin(Z), Z.shape)
mi = (X[x_min, y_min], Y[x_min, y_min], Z.min())
array = np.array([mi])

ax.plot(array[:,0],array[:,1],array[:,2], markerfacecolor = 'black', markeredgecolor = 'black', marker = 'o', markersize = 5, alpha = 1, zorder = 100)
```

```
plt.contour(X, Y, Z, 20, cmap ='winter');
#z_min = np.min(Z)
#cp = ax.contour(X,Y,Z)
#plt.plot(0, 0, 'or', c = "red")
z_min = np.min(Z)
x_min,y_min = np.unravel_index(np.argmin(Z), Z.shape)
mi = (X[x_min, y_min], Y[x_min, y_min], Z.min())
array = np.array([mi])

plt.plot(array[:,0],array[:,1], markerfacecolor = 'black', markeredgecolor = 'black', marker = 'o', markersize = 5, alpha = 1, zorder = 100)
```

###**Function f2 where p2 is a maximum of f2:**

```
def f2(x,y):
  return (2 * x  + 3 * y ** 2)

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f2(X, Y)
```

```
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(X, Y, Z, 50, cmap = 'winter')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
z_max = np.max(Z)
print(z_max)

x_max,y_max = np.unravel_index(np.argmax(Z), Z.shape)
maximum = (X[x_max, y_max], Y[x_max, y_max], Z.max())
array = np.array([maximum])
ax.plot(array[:,0],array[:,1],array[:,2], markerfacecolor = 'black', markeredgecolor = 'black', marker = 'o', markersize = 5, alpha = 1, zorder = 100)
```

###**Function f3 is neither a minimum nor maximum of f3, but the gradient of f3 at p3 is the zero vector:**

```
def f3(x,y):
  return x ** 2 - y ** 2 

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f3(X, Y)
```

```
fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.contour3D(X, Y, Z, 100, cmap = 'Blues')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');

ax.plot([0],[0],[0], markerfacecolor = 'black', markeredgecolor = 'black', marker = 'o', markersize = 5, alpha = 1, zorder = 100)
```
