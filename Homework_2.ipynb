Coverting Sex feature to a new feature called Gender where female=1 and male=0 (From previous homework).

```
# Creates a new dataframe with the feature of 'Sex' replaced with the feature of
# 'Gender', as well as 'male' replaced with '0' and 'female' replaced with '1'
train_df_new.rename({'Sex':'Gender'}, axis=1, inplace=True)
train_df_new['Gender'] = train_df_new['Gender'].replace(['male'],'0')
train_df_new['Gender'] = train_df_new['Gender'].replace(['female'],'1')

test_df_new.rename({'Sex':'Gender'}, axis=1, inplace=True)
test_df_new['Gender'] = test_df_new['Gender'].replace(['male'],'0')
test_df_new['Gender'] = test_df_new['Gender'].replace(['female'],'1')
train_df_new
```

A simple way is to generate random numbers between mean and standard deviation for estimating and completing features with missing or null values. We will first do this for the Age feature(From Homework1):

```
##data_age = [train_df]
# Combines the 'Age' feature of the training and test sets to get the mean and 
# median
age_df = train_df_new['Age'].append(test_df_new['Age'])
# Gets the mean and median of the 'Age' feature of the combined data sets
mean = age_df.mean()
median = age_df.median()

print('Age Mean: ' + str(mean))
print('Age Median: ' + str(median))

# Sets the min and max range of the random number generator
if mean > median:
  min = median
  max = mean
else:
  min = mean
  max = median

# Fills in the null values of the 'Age' feature in the training and test sets
# with random values between the mean and median
for i in range(train_df_new.index.stop):
  if np.isnan(train_df_new.loc[i].at['Age']):
    train_df_new.at[i,'Age'] = np.random.uniform(min, max)

for i in range(test_df_new.index.stop):
  if np.isnan(test_df_new.loc[i].at['Age']):
    test_df_new.at[i,'Age'] = np.random.uniform(min, max)

train_df_new
```

```
# Combines the 'Embarked' feature of training and test sets to get the mode
embarked_df = train_df_new['Embarked'].append(test_df_new['Embarked'])

# Gets the mode of the 'Embarked' feature of the combined data sets
embarked_mode = embarked_df.mode()[0]
print('Embarked Mode: ' + str(embarked_mode))

# Replaces the null values with the mode of the 'Embarked' feature
train_df_new['Embarked'].fillna(embarked_mode, inplace=True)

train_df_new.head(63)
```

```
# Combines the 'Fare' feature of training and test sets to get the mode
fare_df = train_df_new['Fare'].append(test_df_new['Fare'])

# Gets the mode of the 'Fare' feature of the combined data sets
fare_mode = fare_df.mode()[0]
print('Fare Mode: ' + str(fare_mode))

# Replaces the null values with the mode of the 'Fare' feature
test_df_new['Fare'].fillna(fare_mode, inplace=True)

test_df_new
```

Defining a function to calculate the Gini 

```
# Define function to calculate Gini Impurity
def get_gini_impurity(survived_count, total_count):
    survival_prob = survived_count/total_count
    not_survival_prob = (1 - survival_prob)
    random_observation_survived_prob = survival_prob
    random_observation_not_survived_prob = (1 - random_observation_survived_prob)
    mislabelling_survided_prob = not_survival_prob * random_observation_survived_prob
    mislabelling_not_survided_prob = survival_prob * random_observation_not_survived_prob
    gini_impurity = mislabelling_survided_prob + mislabelling_not_survided_prob
    return gini_impurity
```
